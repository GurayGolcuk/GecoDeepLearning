# Convert input data to desired format 

```python
import pickle
import pandas as pd
import numpy as np
from collections import Counter
```

```python
regiondata_url = ".\gmql_data\kirc_data_df.p"
regiondata = pickle.load(open(regiondata_url, "rb"))
```

```python
regiondata
```

Get labels from the index

```python
lbls = regiondata.index.get_level_values('biospecimen_sample__sample_type_id')
len(lbls)
```

```python
lbls= ["01" if x=="05" else x for x in lbls]
for i in range(len(lbls)):
    lbls[i] = int(lbls[i])
```

Dump the unnecessery columns

```python
regiondata = regiondata.normalized_count
regiondata.index = regiondata.index.get_level_values(-1)
regiondata.head()
```

```python
def getFeatures():
    df = pd.read_csv(".\onogenes\kirc.txt", delimiter="\t")
    df = pd.DataFrame(df)
    return list(df.Gene_Symbol)

features = getFeatures()
features.remove("RHEBP1")
features 
```

```python
len(features)
```

```python
regiondata=regiondata[features]
regiondata
```

Shifting the mean value to zero
Makes the variance of each gene equal to one
Fill the NaN values of the REgion data with the mean of its colomns

```python
regiondata = regiondata.fillna(regiondata.mean())
```

```python
regiondata = (regiondata - regiondata.mean())
regiondata = regiondata / (regiondata.max() - regiondata.min())
regiondata
```

Save base data to pickel file

```python
pickle_file = ".\gmql_data\kirc_data_with_label_deneme.p"

try:
    
    f = open(pickle_file, 'wb')
    pickle.dump(regiondata, f, pickle.HIGHEST_PROTOCOL)
    f.close()
except Exception as e:
    print('Unable to save data to', pickle_file, ':', e)
    raise

```

# Split the data for train and test

Since we will use supervised calssification it is logical to do stratified division, since strafied division preserve the data provision ratio

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
pickle_file = ".\gmql_data\kirc_data_with_label_deneme.p"
data = pickle.load(open(pickle_file, "rb"))
data
```

Set labels to y parameter

```python
x = np.array(data)
y = np.array(lbls)
sss = StratifiedShuffleSplit(n_splits=3, test_size=0.2, random_state=0)
sss.get_n_splits(x, y)
sss
```

```python
for train_index, test_index in sss.split(x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

Create validation dataset from training data

```python
x = x_train
z = y_train

sss = StratifiedShuffleSplit(n_splits=3, test_size=0.25, random_state=0)
sss.get_n_splits(x, z)
sss
```

```python
for train_index, valid_index in sss.split(x, z):
    x_train, x_valid = x[train_index], x[valid_index]
    z_train, z_valid = z[train_index], z[valid_index]
```

Save the data for future use

```python
pickle_file = ".\gmql_data\kirc_data_divided_deneme.p"

try:
  f = open(pickle_file, 'wb')
  save = {
    'train_dataset': x_train,
    'train_labels': z_train,
    'test_dataset': x_test,
    'test_labels': y_test,
    'valid_dataset': x_valid,
    'valid_labels': z_valid,
    }
  pickle.dump(save, f, pickle.HIGHEST_PROTOCOL)
  f.close()
except Exception as e:
  print('Unable to save data to', pickle_file, ':', e)
  raise
```
